```python
import requests
import json
import pandas as pd
import numpy as np
```

```python
csv_file='Resources/counties_per_capita_income.csv'
income_counties_df= pd.read_csv(csv_file)
income_counties_df
```

```python
income_counties_df = income_counties_df[['county', 'states', 'pci', 'household_income', 'family_income', 'num_of_households']]

income_counties_df
```

```python
income_counties_df = income_counties_df.rename(columns ={'states': 'state_name', 
'pci': 'per_capita_income'})

income_counties_df
```

```python
income_counties_df['per_capita_income']=income_counties_df['per_capita_income'].str.replace('$','').str.replace(',','')
income_counties_df['household_income']=income_counties_df['household_income'].str.replace('$','').str.replace(',','')
income_counties_df['family_income']=income_counties_df['family_income'].str.replace('$','').str.replace(',','')
income_counties_df['num_of_households']=income_counties_df['num_of_households'].str.replace(',','')

income_counties_df
```

```python
income_counties_df.dtypes
```

```python
income_counties_df['per_capita_income'] = income_counties_df['per_capita_income'].astype('int')
income_counties_df['family_income'] = income_counties_df['family_income'].astype('int')
income_counties_df['household_income'] = income_counties_df['household_income'].astype('int')
income_counties_df['num_of_households'] = income_counties_df['num_of_households'].astype('int')
```

```python
income_counties_df.dtypes
```

```python
income_counties_df = income_counties_df.sort_values(by='household_income', ascending=False)

income_counties_df
```

```python
income_counties_df.index = np.arange(1, len(income_counties_df) + 1)

income_counties_df
```

```python
income_counties_df = income_counties_df.reset_index()
income_counties_df = income_counties_df.rename(columns ={'index': 'household_income_rank'})

income_counties_df
```

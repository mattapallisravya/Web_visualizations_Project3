```python
import sqlite3
import pandas as pd
```

```python
crimes = pd.read_csv('final_merged_dataset.csv')
crimes
```

```python
conn = sqlite3.connect('merged_data.db')
c = conn.cursor()
```

```python
c.execute("""
CREATE TABLE crime_income
(FIPS VARCHAR(10) PRIMARY KEY  NOT NULL,
county VARCHAR(100),
state VARCHAR(50),
crime_rate_per_100000 FLOAT,
crime_rate_rank INTEGER,
murder INTEGER,
rape INTEGER,
robbery INTEGER,
aggravated_assault INTEGER,
burglary INTEGER,
larceny INTEGER,
motor_theft INTEGER,
arson INTEGER,
population_x INTEGER,
total_violent_crime INTEGER,
household_income_rank INTEGER,
state_name VARCHAR(500),
per_capita_income INTEGER,
household_income INTEGER,
family_income INTEGER,
num_of_households INTEGER,
lat FLOAT,
lng FLOAT)""")
```

```python
c.execute('''DROP TABLE IF EXISTS crime_counties''')
```

```python
crimes.to_sql('crime_income', conn, if_exists='append', index = False)
```

```python
c.execute('''SELECT * FROM crime_income''').fetchall()
```

```python
conn.commit()
conn.close()
```

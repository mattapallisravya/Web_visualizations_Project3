```python
import requests
import json
import pandas as pd
```

```python
csv_file='Resources/crime_data_w_population_and_crime_rate.csv'
crime_counties_df= pd.read_csv(csv_file)
crime_counties_df.head()
```

```python
crime_counties_df['state']= crime_counties_df['county_name'].str.split(',').str[1]
# crime_counties_df['county_name']= crime_counties_df['county_name'].str.split(',')

crime_counties_df.head()
```

```python
crime_counties_df["FIPS_ST"] = crime_counties_df.FIPS_ST.map("{:02}".format)
crime_counties_df["FIPS_CTY"] = crime_counties_df.FIPS_CTY.map("{:03}".format)

crime_counties_df.head()
```

```python
crime_counties_df["fips"] = crime_counties_df["FIPS_ST"] + crime_counties_df["FIPS_CTY"]

crime_counties_df.head()
```

```python
crime_counties_df = crime_counties_df[['county_name', 'state', 'crime_rate_per_100000', 
    'index', 'MURDER', 'RAPE', 'ROBBERY', 'AGASSLT', 'BURGLRY', 'LARCENY', 'MVTHEFT', 
    'ARSON', 'population','FIPS_ST', 'FIPS_CTY']]

crime_counties_df
```

```python
crime_counties_df = crime_counties_df.rename(columns ={'index': 'crime_rate_rank', 'MURDER': 'murder', 
'RAPE' : 'rape', 'ROBBERY': 'robbery', 'AGASSLT': 'aggravated_assault', 'BURGLRY': 'burglary', 
'LARCENY': 'larceny', 'MVTHEFT': 'motor_theft', 'ARSON': 'arson'})

crime_counties_df.head()
```

```python
crime_counties_df['total_violent_crime'] = crime_counties_df['murder'] + crime_counties_df['rape'] + crime_counties_df['robbery'] + crime_counties_df['aggravated_assault'] + crime_counties_df['burglary']+ crime_counties_df['larceny'] + crime_counties_df['motor_theft'] + crime_counties_df['arson']

crime_counties_df
```

```python
crime_counties_df['county_name']= crime_counties_df['county_name'].str.split(',').str[0]
crime_counties_df['county_name']= crime_counties_df['county_name'].str.replace('County','')

crime_counties_df.head()
```

```python
crime_counties_df['county_name']= crime_counties_df['county_name'].str.replace('Borough','')
crime_counties_df['county_name']= crime_counties_df['county_name'].str.title()

crime_counties_df
```

```python
crime_counties_df = crime_counties_df.rename(columns ={'county_name':'county'})

crime_counties_df
```

```python
crime_counties_df['county']= crime_counties_df['county'].str.replace('Borough','')

crime_counties_df
```

```python
crime_counties_df.to_csv('complete_crime_counties.csv', index=True)
```

```python
csv_file='complete_income_counties.csv'
income_counties_df= pd.read_csv(csv_file)
income_counties_df.head()
```

```python
# Get list of counties from US census to cross-reference
crime_counties = crime_counties_df[['county','state']]

# Add new column so specify origin of each row after concat
crime_counties.loc[:,"dataset"] = "crime"

# Get list of counties from brewerys dataset to cross-reference
income_counties = income_counties_df[['county','state']]

# Add new column so specify origin of each row after concat
income_counties.loc[:,"dataset"] = "income"

# Combine two dataframes of counties, drop matching counties/state rows from each when they match, sort by state/county
combined_counties = pd.concat([crime_counties, income_counties])\
                        .drop_duplicates(subset=['county', 'state'], keep='first')\
                        .sort_values(['state','county'])

# Drop rows with null values ()
combined_counties.dropna(inplace=True)

# Filter down to scraped county names that do not match census
counties_to_fix = combined_counties[combined_counties['dataset']=='crime']

# Display mismatched county names
print(counties_to_fix)
```

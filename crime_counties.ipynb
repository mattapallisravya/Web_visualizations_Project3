```python
import requests
import json
import pandas as pd
```

```python
csv_file='Resources/crime_data_w_population_and_crime_rate.csv'
crime_counties_df= pd.read_csv(csv_file)
crime_counties_df.head()
```

```python
crime_counties_df['state']= crime_counties_df['county_name'].str.split(',').str[1]
# crime_counties_df['county_name']= crime_counties_df['county_name'].str.split(',')

crime_counties_df.head()
```

```python
crime_counties_df["FIPS_ST"] = crime_counties_df.FIPS_ST.map("{:02}".format)
crime_counties_df["FIPS_CTY"] = crime_counties_df.FIPS_CTY.map("{:03}".format)

crime_counties_df.head()
```

```python
crime_counties_df["fips"] = crime_counties_df["FIPS_ST"] + crime_counties_df["FIPS_CTY"]

crime_counties_df.head()
```

```python
crime_counties_df = crime_counties_df[['county_name', 'state', 'crime_rate_per_100000', 
    'index', 'MURDER', 'RAPE', 'ROBBERY', 'AGASSLT', 'BURGLRY', 'LARCENY', 'MVTHEFT', 
    'ARSON', 'population','FIPS_ST', 'FIPS_CTY']]

crime_counties_df
```

```python
crime_counties_df = crime_counties_df.rename(columns ={'index': 'crime_rate_rank', 'MURDER': 'murder', 
'RAPE' : 'rape', 'ROBBERY': 'robbery', 'AGASSLT': 'aggravated_assault', 'BURGLRY': 'burglary', 
'LARCENY': 'larceny', 'MVTHEFT': 'motor_theft', 'ARSON': 'arson'})

crime_counties_df.head()
```

```python
crime_counties_df['total_violent_crime'] = crime_counties_df['murder'] + crime_counties_df['rape'] + crime_counties_df['robbery'] + crime_counties_df['aggravated_assault'] + crime_counties_df['burglary']+ crime_counties_df['larceny'] + crime_counties_df['motor_theft'] + crime_counties_df['arson']

crime_counties_df
```

```python
crime_counties_df['county_name']= crime_counties_df['county_name'].str.split(',').str[0]
crime_counties_df['county_name']= crime_counties_df['county_name'].str.replace('County','')

crime_counties_df.head()
```

```python
crime_counties_df['county_name']= crime_counties_df['county_name'].str.replace('Borough','')
crime_counties_df['county_name']= crime_counties_df['county_name'].str.title()

crime_counties_df
```

```python
crime_counties_df = crime_counties_df.rename(columns ={'county_name':'county'})

crime_counties_df
```

```python
crime_counties_df['county']= crime_counties_df['county'].str.replace('Borough','')

crime_counties_df
```

```python
crime_counties_df.sort_values(by="county")
```

```python
crime_counties_df.to_csv('complete_crime_counties.csv', index=True)
```

```python
csv_file='complete_income_counties.csv'
income_counties_df= pd.read_csv(csv_file)
income_counties_df
```

```python
# Sorting values by county name
income_counties_df.sort_values(by="county")
```

```python
# Checking dtypes for dataframes
print(income_counties_df.dtypes)
crime_counties_df.dtypes
```

```python
# Stripping empty spaces before and after value in the columns to allow smooth merge
income_counties_df["county"] = income_counties_df["county"].str.strip()
income_counties_df["state"] = income_counties_df["state"].str.strip()
crime_counties_df["state"] = crime_counties_df["state"].str.strip()
crime_counties_df["county"] = crime_counties_df["county"].str.strip()
```

```python
# Merging income dataframe and crime dataframe
merged_crime_income = pd.merge(crime_counties_df,income_counties_df, how = "inner", on=["county","state"])
merged_crime_income
```

```python
# Check column names
merged_crime_income.columns.values
```

```python
# Merging two columns into one to get fips
merged_crime_income["FIPS"] = merged_crime_income["FIPS_ST"] + merged_crime_income["FIPS_CTY"]
merged_crime_income.columns.values
```

```python
# read uscounty csv and rename the column
lat_lng_df= pd.read_csv("uscounties.csv")
lat_lng_df = income_counties_df.rename(columns={"county_fips" : "FIPS"})
lat_lng_df
```

```python
# Checking datatypes before merge
lat_lng_df.dtypes
```

```python
# updated datatype to object
lat_lng_df['FIPS'] = lat_lng_df['FIPS'].astype(str)
lat_lng_df.dtypes
```

```python
# Merging the dataframes to get lat and lng
merged_df = pd.merge(merged_crime_income,lat_lng_df, how = "inner", on = "FIPS")
merged_df
```

```python
merged_df.columns.values
```

```python
merged_df.drop(["FIPS_ST", "FIPS_CTY", "Unnamed: 0", "county_y", "county_ascii", "county_full", "state_id","state_name_y","population_y" ], axis = 1, inplace = True)
```

```python
merged_df.columns.values
```

```python
merged_df.rename(columns = {"county_x":"county", "state_name_x" : "state_name"})
```

```python
# final dataset to push into database
merged_df.to_csv("final_merged_dataset.csv", index=False)
```
